LINEAR REGRESSION MODEL

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import plotly.express as px

# Load data
data_df = pd.read_csv('source_dependent_full_score.csv')

# Features and target
X_features = ['AWL', 'TTR', '-LRB-', '-RRB-', 'AMP', 'ANDC', 'CAUS', 'CC', 'CD', 'COMMA', 'CONC', 'COND', 'CONJ', 'DEMO', 'DEMP', 'DPAR', 'DT', 'DWNT', 'EMPH', 'EX', 'FPP1', 'FW', 'GER', 'HDG', 'IN', 'INPR', 'JJ', 'LS', 'NEMD', 'NN', 'NOMZ', 'OSUB', 'PDT', 'PHC', 'PIN', 'PIT', 'PLACE', 'POMD', 'POS', 'PRED', 'PRMD', 'PRP', 'QUAN', 'QUPR', 'RB', 'RP', 'SPP2', 'SYM', 'SYNE', 'THAC', 'THVC', 'TIME', 'TO', 'TOBJ', 'TPP3', 'TSUB', 'UH', 'VB', 'VBD', 'VBG', 'VBN', 'VPRT', 'WDT', 'WP', 'WPS', 'XX0', '[BEMA]', '[BYPA]', '[CONT]', '[PASS]', '[PASTP]', '[PEAS]', '[PIRE]', '[PRESP]', '[PRIV]', '[PROD]', '[PUBV]', '[SERE]', '[SMP]', '[SPAU]', '[SPIN]', '[STPR]', '[SUAV]', '[THATD]', '[WHCL]', '[WHOBJ]', '[WHQU]', '[WHSUB]', '[WZPAST]', '[WZPRES]', '_NN']

print(f"Number of features in X_features: {len(X_features)}")
X = data_df[X_features]  # Features
y = data_df['mean_score (actual/max*100)']  # Target variable for score
print(f"Columns in X: {X.columns.tolist()}")

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Model and fit
lin_reg = LinearRegression()
lin_reg_model = lin_reg.fit(X_train, y_train)

# Predictions
y_pred = lin_reg_model.predict(X_test)

# Metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = metrics.mean_squared_error(y_test, y_pred, squared=False)
r2 = metrics.r2_score(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2): {r2}')

# Save metrics to a text file
with open('LinearRegression_metrics.txt', 'w') as f:
    f.write(f'Mean Absolute Error (MAE): {mae}\n')
    f.write(f'Mean Squared Error (MSE): {mse}\n')
    f.write(f'Root Mean Squared Error (RMSE): {rmse}\n')
    f.write(f'R-squared (R2): {r2}\n')

# Feature importance (coefficients)
feature_importance = lin_reg_model.coef_
feature_importance_df = pd.DataFrame(feature_importance, index=X_features, columns=['Coefficient'])
feature_importance_df.to_csv('LinearRegression_feature_importance.csv')

# Scatter plot of predicted vs actual scores
fig = px.scatter(x=y_test, y=y_pred, labels={'x': 'Actual Score', 'y': 'Predicted Score'}, title='Actual vs Predicted Scores')
fig.add_shape(type='line', x0=min(y_test), y0=min(y_test), x1=max(y_test), y1=max(y_test), line=dict(color='red',))
fig.show()


```
